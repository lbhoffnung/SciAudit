# 🛡️ SciAudit Demo: Integridade Científica em Notebooks

Este notebook demonstra como o **SciAudit** detecta violações metodológicas críticas em fluxos de trabalho de Data Science comuns.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

df = pd.DataFrame({
    'feature_a': np.random.rand(100),
    'feature_b': np.random.rand(100),
    'target': np.random.randint(0, 2, 100)
})

# 🔴 SCI-008: Label Leakage
df['helper_feature'] = df['target'] * 2

X = df.drop('target', axis=1)
y = df['target']
```

### ⚠️ O Pecado da Reprodutibilidade
Note que não passamos `random_state`. Isso gera resultados diferentes a cada execução.

```python
# 🟡 SCI-002: Falta de Random Seed
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

### ⚠️ Contaminação de Teste
Abaixo, o modelo é treinado usando o conjunto de teste por engano.

```python
# 🔴 SCI-006: Test Set Contamination
model = RandomForestClassifier() # Outro SCI-002 aqui
model.fit(X_test, y_test) 
```

### 📊 Resultado Ilusório
Sem validação cruzada, o score parece ótimo, mas é inválido.

```python
# 🟡 SCI-005: Overfitting Cego
acc = accuracy_score(y_test, model.predict(X_test))
print(f"Acurácia: {acc}")
```
